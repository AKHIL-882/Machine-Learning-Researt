```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")
```

```python
# find no.of data points
print(dataset.shape)
```

```python
# find columns
print(dataset.columns)
```

```python
# find data points for each columns
dataset['variety'].value_counts()
```

```python
# 2d scatter plot

dataset.plot(kind='scatter',x='sepal.length',y='sepal.width')
plt.show
```

```python

dataset.plot(kind='scatter',x='petal.length',y='petal.width')
plt.show
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(dataset,hue="variety",size=4) \
   .map(plt.scatter,"sepal.length","sepal.width") \
   .add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(dataset,hue="variety",size=4)\
   .map(plt.scatter,"petal.length","petal.width")\
   .add_legend()
plt.show()
```
